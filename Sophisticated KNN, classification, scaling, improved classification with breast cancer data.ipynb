```python
import pandas as pd
import numpy as np

# import plotting libraries
import matplotlib.pyplot as plt

# import the ML algorithm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("breast-cancer-diag.csv")
```

```python
df.info()
```

```python
df.columns
```

```python
#Number of rows and columns 
df.shape
```

```python
#Null values
df.isnull().sum()
```

```python
#Dropping ID and Unnamed since they are not useful and there are no missing values - axis = 1 to drop a column, it is 0 when dropping a row
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
#Now there is one less column
df.shape
```

```python
x = df.drop(['diagnosis'], axis=1)
y = df.diagnosis.values
```

```python
x
```

```python
#Malignant and benign
y
```

```python
#Instantiate the KNN Model 
knn = KNeighborsClassifier()
```

```python
#Train the model (fit)
knn.fit(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, shuffle=True)
```

```python
#Train vs. test sampling
x_train.shape, x_test.shape
```

```python
#Lazy analysis - Remember, there's nothing to fit in cluster analysis!
knn.fit(x_train, y_train)
```

```python
###Predict the test samples 
y_pred = knn.predict(x_test)
```

```python
###Evaluate the model outcomes using classification metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#AUC score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#Confusion matrix
print(classification_report(y_test, y_pred))
```

```python
scaler.fit(x_val)
rescaled_x = scaler.transform(x_val)
```

```python
x_train, x_test, y_train, y_test = train_test_split(rescaled_x,y, test_size=.25, shuffle=True)
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
```

```python
#Auc has improved
accuracy_score(y_test, y_pred)
```

```python
#Confusion matrix has improved
confusion_matrix(y_test,y_pred)
```

```python
#Classification report has improved!
print(classification_report(y_test,y_pred))
```
